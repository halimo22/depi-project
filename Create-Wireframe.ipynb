```python
import matplotlib.pyplot as plt

def draw_wireframe():
    fig, ax = plt.subplots(figsize=(6, 12))
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 12)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_frame_on(False)

    # Header (Search Bar)
    ax.add_patch(plt.Rectangle((0, 11), 12, 0.7, fill=True, color='lightgray', edgecolor='black'))
    ax.text(3, 11.35, "Search Movies...", fontsize=10, ha='center', va='center')

    # Recommendation Section
    ax.text(0.5, 10.75, "Recommended for You", fontsize=12, fontweight='bold', ha='left', va='center')
    for i in range(3):
        ax.add_patch(plt.Rectangle((0.5 + i * 4, 8.5), 3, 2, fill=False, edgecolor='black'))
        ax.text(0.5 + i * 4 + 0.75, 8.25, "Movie " + str(i+1), fontsize=8, ha='center')

    # Categories
    ax.text(0.5, 7.5, "Categories", fontsize=12, fontweight='bold', ha='left', va='center')
    for i in range(2):
        for j in range(2):
            ax.add_patch(plt.Rectangle((0.5 + j * 6, 6 - i * 1.5), 5, 1.2, fill=False, edgecolor='black'))
            ax.text(3 + j * 6, 6 - i * 1.5 + 0.6, "Category " + str(i * 2 + j + 1), fontsize=8, ha='center')

    # Popular Movies Section
    ax.text(0.5, 3.85, "Popular Movies", fontsize=12, fontweight='bold', ha='left', va='center')
    for i in range(3):
        ax.add_patch(plt.Rectangle((0.5 + i * 4, 1.5), 3, 2, fill=False, edgecolor='black'))
        ax.text(1.25 + i * 4 , 1.25, "Movie " + str(i+4), fontsize=8, ha='center')

    # Bottom Navigation
    ax.add_patch(plt.Rectangle((0, 0), 12, 0.8, fill=True, color='lightgray', edgecolor='black'))
    for i, label in enumerate(["Home", "Browse", "Profile"]):
        ax.text(2 + i * 4, 0.4, label, fontsize=10, ha='center', va='center')

    plt.show()

draw_wireframe()
```
